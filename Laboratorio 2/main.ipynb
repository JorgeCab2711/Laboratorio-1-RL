En este escenario el agente puede tomar ciertas acciones en distintos estados , lo que provoca que obtenga una recompensa del ambiente, no obtenga nada o directamente oibtenga un castigo. 

2. Los componentes principales serían el agente y el ambiente. Podría decirse que los estados, recompensas y actores son factores secundarios.

3. El objetivo principal del aprendizaje por refuerzo como son las Cadenas de Markov sería maximizar el nivel de recompensa del agente para su aprendizaje, a maximizar no me refiero a llegar a la recompensa máxima, si no a obtener la mayor cantidad de recompensa posible.
